```
df=pd.read_csv("car-sales-extended-missing-data.csv")
df
```

```
df.isna().sum()
```

```
df['Make']=df['Make'].fillna(df['Make'].mode()[0])
df['Colour']=df['Colour'].fillna(df['Colour'].mode()[0])
df['Odometer (KM)']=df['Odometer (KM)'].fillna(df['Odometer (KM)'].median())
df['Doors']=df['Doors'].fillna(df['Doors'].median())
df['Price']=df['Price'].fillna(df['Price'].median())
```

```
df
```

```
df['price-per-km']=df['Price'] / df['Odometer (KM)']
```
