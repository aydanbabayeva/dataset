```
import pandas as pd

url = 'https://raw.githubusercontent.com/aydanbabayeva/dataset/refs/heads/main/healthcare-dataset-stroke-data.csv'  # Ã¶z URL-in
df = pd.read_csv(url)
df.head()

```

```
#clean eden id sutunu modele lazim deyil deyek ve silek
df=df.drop(columns=["id"])
```

```
df.columns
```

```
#nan
df.isnull().sum()
```

```
df=df["bmi"].fillna(df["bmi"].median())
```

```
df.isnull().sum()
```

```
#Categorical Encoding
df=pd.get_dummies(df, drop_first=True)
```

```
#scaling normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_cols = ["age", "avg_glucose_level", "bmi"]
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```
#outlier
Q1 = df["avg_glucose_level"].quantile(0.25)
Q3 = df["avg_glucose_level"].quantile(0.75)
IQR = Q3 - Q1
```

```
df_clean = df[(df["avg_glucose_level"] > Q1 - 1.5*IQR) & (df["avg_glucose_level"] < Q3 + 1.5*IQR)]
```
